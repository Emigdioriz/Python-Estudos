## Criando um modelo de predição de alugueis no Rio de Janeiro

## Importando as bibliotecas

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import statsmodels.api as sm
```

### Importando o dataset já organizado

```python
df = pd.read_csv('dataset.csv')
df
```

### Retirando as colunas log_Valor e Bairro do dataset, para que possamos usá-lo como X

Porém, primeiro, vamos indexar a coluna Bairro a uma variável dummy para que possamos fazer a transformação mais pra frente utilizando o OneHotEncoder

```python
dummy = df.Bairro
X_cru = df.drop(['Bairro','log_Valor'],axis=1)
X_cru.fillna(0, inplace=True)
X_cru
```

```python
dummy
```

Transformando a series dummy em um dataframe

```python
dummy = pd.DataFrame(dummy)
dummy
```

Criando a tabela de variáveis dummies

```python
encoder = OneHotEncoder(sparse=False, drop='first')
variavel_categorica_encoded = encoder.fit_transform(dummy)
variavel_categorica_encoded
```

```python
df_encoded = pd.DataFrame(variavel_categorica_encoded, columns=encoder.get_feature_names_out(input_features=['Bairro']))
df_encoded
```

### Agora vamos unir o DataFrame X_cru com o dataFrame df_enconded para que possamos formar nosso X

```python
X_cru.reset_index(drop=True, inplace=True)
df_encoded.reset_index(drop=True, inplace=True)
X = pd.concat([X_cru, df_encoded], axis=1)
```

Nota-se que a dimensão de colunas de X_cru + df_encoded tem q ser igual a dimensão de colunas de X

```python
X_cru.shape[1] + df_encoded.shape[1] == X.shape[1]
```

```python
X
```

## Criando nosso modelo de regressão 

Devido a problemas ja encontrados nos valores, precisamos corrigir valores de log_valor que sejam infinitos no dataFrame df

```python
y = df.log_Valor
y = y.replace([np.inf, -np.inf], 0)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=10)
```

### Usando a biblioteca statsmodels, faremos uma avaliação qualitativa do nosso modelo, antes de fazermos as simulações

```python
X_train_com_constante = sm.add_constant(X_train)
X_train_com_constante
```

```python
modelo_statsmodels = sm.OLS(y_train, X_train_com_constante, hasconst = True).fit()
```

Vamos avaliar a lista de resultados qualitativos

```python
modelo_statsmodels.summary()
```

<font color='red' style='font-size: 15px;'>**F-Statistic** - Teste de significäncia = testa a significância dos parâmetros do modelo conjuntamente. Se o valor de **Prob (F-statistic)** for acima de 0.05, devemos rejeitar o modelo por não ser estatisticamente relevante. Significa que os parâmetros não possuem uma boa relação.</font> 

<font color='red' style='font-size: 15px;'>**Analisando o teste individual dos parâmetros**
Analisando os valores de **t** e de **P>|t|**: vale a mesma regra do 0.05. 
Nota-se que nesse modelo o parâmetro **log_Quartos** tem um probabilidade de t de 0.411. Sendo estatisticamente insignificante para este modelo. </font> 

<hr style='border: 2px solid red;'>

##  Teste de significância conjunta dos parâmetros

  <font color='black' style='font-size: 15px;'>Prob(F-statistic) < 0.05 (**ok**)</font>


##  Teste de significância individual dos parâmetros

   <font color='black' style='font-size: 15px;'>P>|t| < 0.05 (**ok**)</font>


# <font color='red' style='font-size: 30px;'>Avaliando o resultado da tabela qualitativa</font>
<hr style='border: 2px solid red;'>

Tivemos um **Prob (F-statistic)** muito bom, porém alguns bairros e a variável de quartos provaram-se ser de **baixa relevância** para nosso modelo.

Será necessário removê-los do modelo e realizar o teste qualitativo novamente 

## Extraindo de X as variáveis que possuem P>|t| maiores que 0.05.

```python
summary = modelo_statsmodels.summary()
results = summary.tables[1]
results = summary.tables[1].as_html()
results = pd.read_html(results, header=0, index_col=0)[0]
results['P>|t|'] = pd.to_numeric(results['P>|t|'])
variaveis_selecionadas = results[results['P>|t|'] > 0.05]
```

```python
variaveis_selecionadas
```

Recriando-as do nosso X

```python
nomes_variaveis_selecionadas = variaveis_selecionadas.index.tolist()
X = X.drop(nomes_variaveis_selecionadas, axis=1)
X
```

## Recriando o modelo com as variáveis filtrada

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2811)
X_train_com_constante = sm.add_constant(X_train)
modelo_statsmodels = sm.OLS(y_train, X_train_com_constante, hasconst = True).fit()
```

```python
modelo_statsmodels.summary()
```

### Agora temos um modelo que se enquandra dentro dos padrões de qualidade. Podemos começar a montar a previsões

```python
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
print('R² = {}'.format(modelo.score(X_train, y_train).round(3)))
```

```python
y_previsto = modelo.predict(X_test)
print('R² = %s' % metrics.r2_score(y_test, y_previsto).round(3))
```

## Previsões pontuais

Vamos criar um print dos bairros disponíveis para a simulação, pois durante o filtro de qualidade alguns bairros foram retirados da simulação

```python
bairros = X.columns[5:]
```

```python
for bairro in bairros:
    print(bairro)
```

## Estimando o erro

```python
y_test_real = np.exp(y_test)
y_previsto_real = np.exp(y_previsto)

EQM = metrics.mean_squared_error(y_test_real, y_previsto_real).round(2)
REQM = np.sqrt(metrics.mean_squared_error(y_test_real, y_previsto_real)).round(2)
R2 = metrics.r2_score(y_test_real, y_previsto_real).round(2)


metricas = pd.DataFrame([EQM, REQM], ['EQM', 'REQM'], columns=['Métricas'])
metricas
```

## Dados para a simulação

```python
bairro_simulado = 'Bairro_Copacabana'
Area_apartamento = 240
Vagas_apartamento = 2
Suites = 3
Condominio = 1500
IPTU = 750


# Criando a simulação 
simulacao = pd.DataFrame({
    'Area': [Area_apartamento],
    'Vagas': [Vagas_apartamento],
    'Suites': [Suites],
    'Condominio': [Condominio],
    'IPTU': [500]
})


# Transformando os valores definidos para a simulação para valores em logarítmo, para que a modelo funcione

valores_log = pd.DataFrame()
for chave, valor in simulacao.items():
    if chave not in bairros:  # Não aplique log nas variáveis dummy
        valores_log['log_' + chave] = np.log(valor + 1)
        
# Crie as variáveis dummy automaticamente com base no bairro simulado
list=[]
for bairro in bairros:
    if bairro == bairro_simulado:
        valores_log[bairro] = 1    
    else:
        valores_log[bairro] = 0
        
# Agora que temos o dataframe com os dados desejados para a simulaçao, é so aplicarmos esses dados ao modelo

print('Preço estimado de aluguel para %s é de R$%s' % (bairro_simulado,np.exp(modelo.predict(valores_log)[0]).round(2)))
print('Erro médio no valor do aluguel é de R${}'.format(metricas.iloc[1,0]  ))
```
